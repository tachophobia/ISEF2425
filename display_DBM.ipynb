```python
from DBM import DBM
import matplotlib.pyplot as plt
```

```python
dim=200
dbm = DBM(eta=8, dim=dim, seed=(0, dim))

dbm.add_circle_of_charge((dim+80, dim+100), 20, 1)
dbm.add_line_of_charge((dim+100, dim+100), (dim+150, dim+100), 1)
dbm.add_line_of_charge((dim+150, dim+100), (dim*2, dim+80), 1)
dbm.add_line_of_charge((dim+150, dim+100), (dim*2, dim+120), 1)
dbm.add_line_of_charge((dim+100, dim+100), (dim+140, dim+80), 1)
dbm.add_line_of_charge((dim+100, dim+100), (dim+140, dim+120), 1)


steps, frames = dbm.simulate(steps=5000, animate=True)
dbm.show()
```

```python
import matplotlib.animation as animation
import numpy as np

dpi = 100
fig, ax = plt.subplots(figsize=(10, 10))
fig.tight_layout(pad=0)
def update(frame):
    ax.clear()
    ax.axis('off')
    frame[np.where(frame<0)] = 0
    ax.imshow(frame, cmap='afmhot')

f = []
i = 1
n = 3
while i < len(frames):
    f.append(frames[int(i)])
    i = i * n
    n = max(1.05, n*0.9)
    

ani = animation.FuncAnimation(fig, update, frames=f, interval=1)
ani.save("movie.gif", dpi=dpi)
```
