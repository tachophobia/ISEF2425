```python
from DBM import DBM
import numpy as np
```

```python
dbm = DBM(eta=100, dim=100, seed=(0, 100))
dbm.add_line_of_charge((2*dbm.dim, 0), (2*dbm.dim, 2*dbm.dim), 1e5)
dbm.add_rectangle_of_charge((1.5*dbm.dim, 1.5*dbm.dim), (2*dbm.dim, 1.7*dbm.dim), 1e5)
dbm.simulate(steps=1000)
dbm.show(show_environment=True)
dbm.show()
```
